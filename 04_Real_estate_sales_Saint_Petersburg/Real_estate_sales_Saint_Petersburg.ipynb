# Исследование объявлений о продаже квартир

## Описание проекта

Источник данных — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. 

Вида данных:
- получены от пользователя
- получены автоматически на основе картографических данных

## Цель исследования

Определение факторов, от которых зависит цена квартиры

Проверка гипотез:
- Гипотеза 1: квартиры в центре Санкт-Петербурга не отличаются от квартир с другим месторасположением
- Гипотеза 2: факторы, влияющие на цену квартиры, не зависят от ее месторасположения

## Ход исследования

1. Предварительная обработка данных:
- выбор данных
- изменение заголовков столбцов
- определение и заполнение пропущенных значений
- удаление дубликатов
- изменение типа данных
2. Расчет показателей:
- цена квадратного метра
- день недели, месяц и год публикации объявления
- этаж квартиры
- оотношение жилой и общей площади
- отношение площади кухни к общей
3. Анализ данных и визуализация данных
- площадь
- цена
- число комнат
- высота потолков 
- время продажи квартиры
4. Определение факторов, влияющих на стоимость квартиры:
- площадь
- число комнат
- удалённость от центра
- этаж 
- дата размещения: день недели, месяц, год
5. Нахождение  10 населённых пунктов с наибольшим числом объявлений:
- расчет средней цены квадратного метра в этих населённых пунктах.
6. Анализ квартир в центре Санкт-Петербурге
- определение центра Санкт-Петербурга
- анализ параметров: площадь, цена, число комнат, высота потолков
- факторы, влияющие на стоимость квартиры 
8. Общие выводы

```python
from IPython.core.display import display


def observe_data(table):
    '''Giving common understanding of data in dataframe'''
    print ('Случайная запись таблицы')
    display(table.sample())
    print('_'*100)
    print(f'Количество записей в таблице: {table.shape[0]} \nKоличество столбцов в таблице: {table.shape[1]}')
    print('_'*100)
    names_columns = table.columns
    print(f'Названия столбцов: {names_columns}')
    print('_'*100)
    names_to_corrige = []
    for name_column in names_columns:
        if not name_column.islower() or ' ' in name_column:
            names_to_corrige.append(name_column)
    if len(names_to_corrige) == 0:
        print ('Название столбцов корректно')
    else:
        print(f'Название следующих колонок должны быть откорретированы: {names_to_corrige}')
    print('_'*100)
    lost_values = table.isna().sum()
    missed_values = lost_values.to_frame(name='missed_values')
    missed_values['percent'] = round(table.isna().mean()*100, 0)
    print ('Пропущенные значения в абсолютном и относительном выражениях:')
    display(missed_values)
    print('_'*100)
    print('Типы данных в колонках:')
    print(table.dtypes)
    print('_'*100)
    print(f'Количество явных дубликатов: {table.duplicated().sum()}', 
          f'что составляет {round(table.duplicated().mean()*100, 0)} % от всех записей')
observe_data(df)
```

#### Выбросы

Выбросы для площади с учетом ее независимости от других признаков

```python
def delete_extreme_outliers (table, column, percent_of_outliers = 5, method_to_delete = 'quantile'):
    """function for counting and deleting outliers according choosen method"""
    #check outliers by standart deviation
    fig, ax =plt.subplots(1,2)
    sns.histplot (data = table, x = column, ax=ax[0] )
    sns.boxplot (data = table, x = column, ax=ax[1])
    fig.show()
    lower_bound_hist = table[column].mean()-3*table[column].std()
    upper_bound_hist = table[column].mean()+3*table[column].std()
    #check outliers by boxplot - moderate and extreme
    IQR = table[column].quantile(q=0.75) - table[column].quantile(q=0.25)
    lower_bound_boxplot_moderate = 2.5*table[column].quantile(0.25)-1.5*table[column].quantile(q=0.75)
    upper_bound_boxplot_moderate = 2.5*table[column].quantile(0.75)-1.5*table[column].quantile(q=0.25)

    lower_bound_boxplot_extreme = 4*table[column].quantile(0.25)-3*table[column].quantile(q=0.75)
    upper_bound_boxplot_extreme = 4*table[column].quantile(0.75)-3*table[column].quantile(q=0.25)
    
    #check outliers by quantiles
    lower_bound_quantile = table[column].quantile(q = percent_of_outliers/2/100)
    upper_bound_quantile = table[column].quantile(q = 1 - percent_of_outliers/2/100)
    
    if method_to_delete == 'quantile':
        lower_bound = lower_bound_quantile
        upper_bound = upper_bound_quantile
    elif method_to_delete == 'boxplot_extreme':
        lower_bound = lower_bound_boxplot_extreme
        upper_bound = upper_bound_boxplot_extreme
    elif method_to_delete == 'boxplot_moderate':
        lower_bound = lower_bound_boxplot_moderate
        upper_bound = upper_bound_boxplot_moderate
    else:
        lower_bound = lower_bound_hist
        upper_bound = upper_bound_hist
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f'Percent of outliers  {round(100*len(outliers)/len(table), 0)}')
    table = table.drop(labels = outliers.index)
    print(f'Outliers according {method_to_delete} < {lower_bound.round(2)} and > {upper_bound.round(2)}')
    print(f'Rows after deleting {len(table)}. Method {method_to_delete}')
    return table, outliers
```

```python
def delete_extreme_outliers_cat(table, cols, cols_for_grouping):
    for col in cols:
        for name, grouped_data in table.groupby(cols_for_grouping):
            lower_bound_boxplot_extreme = 4*grouped_data[col].quantile(0.25)-3*grouped_data[col].quantile(0.75)
            upper_bound_boxplot_extreme = 4*grouped_data[col].quantile(0.75)-3*grouped_data[col].quantile(0.25)
            print(name, lower_bound_boxplot_extreme, upper_bound_boxplot_extreme)
            outliers = grouped_data[(grouped_data[col] < lower_bound_boxplot_extreme) | (grouped_data[col] > upper_bound_boxplot_extreme)]
            table = table.drop(labels = outliers.index)
        print(f'Количество удаленных записей: {len(outliers)}')
    return table, outliers
```

```python
#function for drawing 2 hystogrammes in the same coordinate systeme
def draw_two_hist (table1, table2, col, title, ox, oy, xtick):
    plt.hist(table1[col],  alpha = 0.5, label='all cities') 
    plt.hist(table2[col],  alpha = 0.5, label='Saint-Petersburg')
    plt.legend(loc='upper right')
    plt.xlabel(ox)
    plt.ylabel(oy)
    plt.title(title)
    plt.xticks(range(0, table1[col].max().astype('int'), xtick))
```

```python
#function for drawing two bars in same coordinate systeme
def draw_two_bars(table1, table2, col1, col2):
    pivot1 = table1.pivot_table(index = col1, values = col2, aggfunc = 'median')
    pivot2 = table2.pivot_table(index = col1, values = col2, aggfunc = 'median')
    res = pivot1.merge(pivot2, right_index = True, left_index = True, suffixes=('_general', '_piter'))
    res.plot(kind = 'bar')
    plt.ylabel('price square metre, thousand rub')
    plt.ylim(60, 200)
    print(f'Средняя стоимость квадратного метра в центре {round(piter_centre["price_square_metre"].median(), 0)}')
```

```python
draw_two_bars(df.query('city_centers_nearest >= 9000'), piter_centre, 'type_floor', 'price_square_metre')
plt.title('Этаж оказывает одинаковое влияние на стоимость квадратного метра в центре и не в центре')
```
